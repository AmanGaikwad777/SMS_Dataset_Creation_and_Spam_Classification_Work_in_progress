```python
# 3. Load dataset
df = pd.read_csv(r"C:\Users\amang\sample_project\Rishabh_mess.csv")  # Ensure column: "message_body"
```

```python
# 4. Define label categories
label_categories = [
    "Phishing",
    "Smishing",
    "Promotional",
    "Loan/Financial Scam",
    "Job Scam",
    "Crypto/Investment Scam",
    "No Spam"
]

labels_str = ", ".join(label_categories)
```

```python
# 5. Define a function to classify a single SMS (new API)
def classify_sms_gpt(text):
    prompt = f"""
    Classify the following SMS into one of the categories: {labels_str}.
    SMS: \"{text}\"
    Only respond with the category name exactly as one of the options above.
    """
    try:
        response = openai.chat.completions.create(
            model="gpt-4o-mini",
            messages=[{"role": "user", "content": prompt}],
            temperature=0
        )
        label = response.choices[0].message.content.strip()
        if label not in label_categories:
            label = "No Spam"
        return label
    except Exception as e:
        print("Error:", e)
        return "No Spam"
```

```python
# 6. Loop over dataset and classify
predicted_labels = []

for idx, row in df.iterrows():
    text = str(row["message_body"])
    label = classify_sms_gpt(text)
    predicted_labels.append(label)
    time.sleep(0.5)  # small delay to avoid hitting rate limits
```

```python
# 7. Save annotated dataset
df["predicted_label"] = predicted_labels
df.to_csv(r"C:\Users\amang\sample_project\annotated_dataset.csv", index=False)
print("Annotated dataset saved as annotated_dataset.csv")
```
